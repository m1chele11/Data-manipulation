# Part 0:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
file = 'Fish.csv'
data = pd.read_csv(file)

print(data.head())
```

```python
columns = data.columns.tolist()
print("Columns in the dataset:", columns)

numeric_data = data.select_dtypes(include=[np.number])

ranges = numeric_data.max() - numeric_data.min()
print("Range of values for each measurement:")
print(ranges)


```

```python
max_range_measurement = ranges.idxmax()
max_range_value = ranges.max()
```

# Part 1:

```python
# 1
numeric_data = data.select_dtypes(include=[np.number])
numeric_columns = numeric_data.columns.difference(['Weight'])  

plt.figure(figsize=(15, 10))
for column in numeric_columns:
    plt.scatter(data[column], data['Weight'], label=column)
    plt.title(f'Scatter Plot of {column} vs. Weight')
    plt.xlabel(column)
    plt.ylabel('Weight')
    plt.grid()
    plt.legend()
    plt.show()
```

2:
- The separation into distinct curves could be due to biological differences among fish species. Different species may have unique growth patterns and body shapes that influence how measurements like length, height, and width relate to weight. For example, some species may have a more elongated body shape while others might be stockier.

3: 
- Yes, it looks like the higher in length one is, then the higher in weight they will also be. 

# Part 2:

```python
#1
X = data['Length2']
Y = data['Weight']


slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)


print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")
```

```python
# 2
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')

Y_pred = slope * X + intercept

plt.plot(X, Y_pred, color='red', label='Regression Line')

plt.title('Linear Regression: Length2 vs. Weight')
plt.xlabel('Length2')
plt.ylabel('Weight')
plt.legend()
plt.grid()
plt.show()
```

3:
- The regression line provides a general trend but does not perfectly fit the data points. There are areas where the line consistently overestimates weights, particularly at lower lengths, and underestimates weights at higher lengths. This pattern may result from biological variability among species, indicating that the relationship between length and weight is not purely linear. Factors such as growth rates and measurement errors could also contribute to these deviations. Overall, while the model gives a basic understanding of the relationship, further analysis or a more complex model may be necessary for better accuracy.

# Part 3:

```python
#1
data['Estimated Volume'] = data['Length2'] * data['Width'] * data['Height']
print(data)
```

```python
X_volume = data['Estimated Volume']
Y_weight = data['Weight']


slope_volume, intercept_volume, r_value_volume, p_value_volume, std_err_volume = stats.linregress(X_volume, Y_weight)


print(f"Slope (m): {slope_volume}")
print(f"Intercept (b): {intercept_volume}")
```

```python
#2
plt.figure(figsize=(10, 6))
plt.scatter(X_volume, Y_weight, color='blue', label='Data Points')

Y_pred_volume = slope_volume * X_volume + intercept_volume

plt.plot(X_volume, Y_pred_volume, color='red', label='Regression Line')


plt.title('Linear Regression: Estimated Volume vs. Weight')
plt.xlabel('Estimated Volume')
plt.ylabel('Weight')
plt.legend()
plt.grid()
plt.show()
```

3:
- Yes, the line now more closley resembles the data. Although there are still some outliers we see that the regression line has closer clustering of the points. This shows the reduced overestimation and underestimation of the model.

4:
- Understanding that weight is more closely related to volume can have practical implications in fields such as fisheries science and aquaculture, where predicting fish weight based on physical dimensions is crucial for management and sustainability practices.

```python


data['Estimted Vlume'] = data['Length2'] * data['Width'] * data['Height']

plt.figure(figsize=(12, 8))

for species in data['Species'].unique():
    species_data = data[data['Species'] == species]
    
    X_volume = species_data['Estimated Volume']
    Y_weight = species_data['Weight']
    
    slope, intercept, r_value, p_value, std_err = stats.linregress(X_volume, Y_weight)
    
    Y_pred = slope * X_volume + intercept
    plt.scatter(X_volume, Y_weight, label=f'{species} Data', alpha=0.5)
    plt.plot(X_volume, Y_pred, label=f'{species} Regression', linewidth=2)

plt.title('Linear Regresion of Estimated Weight by Species')
plt.xlabel('Estimated Volume')
plt.ylabel('Weight')
plt.legend()
plt.grid()
plt.show()
```

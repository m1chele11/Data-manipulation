```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Part 1:




```python
#1:
n = 10 

#2
dice_rolls = np.random.randint(1, 7, size=(n, 2))
print(dice_rolls)

#4 - 10:
# Calculate the required values
sums = dice_rolls.sum(axis=1)
products = dice_rolls.prod(axis=1)
max_values = dice_rolls.max(axis=1)
min_values = dice_rolls.min(axis=1)

# Create the DataFrame
df = pd.DataFrame({
    'Dice Pair': [tuple(roll) for roll in dice_rolls],
    'Sum': sums,
    'Product': products,
    'Max': max_values,
    'Min': min_values
})

print(df)
```

```python
#with Tuple:

dice_rolls = np.random.randint(1, 7, size=(n, 2))
print(dice_rolls)


data = []
for roll in dice_rolls:
    pair = tuple(roll)
    sum_roll = sum(pair)
    product_roll = pair[0] * pair[1]
    max_roll = max(pair)
    min_roll = min(pair)
    data.append((pair, sum_roll, product_roll, max_roll, min_roll))

#DataFrame from the list of tuples
df_tuples = pd.DataFrame(data, columns=['Dice Pair', 'Sum', 'Product', 'Max', 'Min'])
print(df_tuples)


#the tuple df looks the same as my array df. I might have done it wrong. But I think it the same since the arrays are converted to tuples in the first one.
```

# Part 2:

```python
# Histogram for Sum
plt.subplot(2, 2, 1)
plt.hist(df_tuples['Sum'], bins=np.arange(2, 14) - 0.5, edgecolor='black')  # Align ticks
plt.title('Distribution of Sum')
plt.xlabel('Sum')
plt.ylabel('Frequency')
plt.xticks(range(2, 13));
```

```python
# Histogram for Product
plt.subplot(2, 2, 2)
plt.hist(df_tuples['Product'], bins=np.arange(1, 37) - 0.5, edgecolor='black')
plt.title('Distribution of Product')
plt.xlabel('Product')
plt.ylabel('Frequency')
plt.xticks(range(1, 37, 3));
```

```python
# Histogram for Max
plt.subplot(2, 2, 3)
plt.hist(df_tuples['Max'], bins=np.arange(1, 8) - 0.5, edgecolor='black')
plt.title('Distribution of Max')
plt.xlabel('Max')
plt.ylabel('Frequency')
plt.xticks(range(1, 7));
```

```python
# Histogram for Min
plt.subplot(2, 2, 4)
plt.hist(df_tuples['Min'], bins=np.arange(1, 8) - 0.5, edgecolor='black')
plt.xlabel('Min')
plt.ylabel('Frequency')
plt.xticks(range(1, 7));
```

```python
#DF.hist :
# Using the DataFrame's built-in hist method
df_tuples[['Sum', 'Product', 'Max', 'Min']].hist(bins=10, edgecolor='black', figsize=(12, 10), layout=(2, 2))
plt.suptitle('Distribution of Columns using DataFrame hist()')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
#2: 
from scipy.stats import norm
#3:
df = pd.read_csv('CompanyABCProfit.csv')
print(df)

#4:
print(df)
num_rows = df.shape[0]

print(f'The DataFrame has {num_rows} rows.') 
#It has 200 rows
```

# Part 2:

```python
#1:
profit_data = df['Profit(Rs \'000)'].dropna()
# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(profit_data, bins=20, density=True, alpha=0.6, color='g', edgecolor='black');

#2:
#I would exect the distribution to be normal since it is a companies profit data, unless the company is like bankrupt or rising in profits very fast.

```

```python
#3
mu, std = norm.fit(profit_data)

# Generate a range of values for the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plot the normal distribution
plt.plot(x, p, 'k', linewidth=2)
title = "plotted"
plt.title(title)
plt.xlabel('Profit (Rs \'000)')
plt.ylabel('Density')
plt.grid()
plt.show()
```
